```python
M = pm.MCMC(five_species_model)
```

```python
M.sample(iter=1000, burn=100, thin=10)
```

```python
M_MAP = pm.MAP(five_species_model)
M_MAP.fit()
```

```python
from pylab import hist, show

hist(M.trace('theta_B_CEAD')[:])
show()
```
